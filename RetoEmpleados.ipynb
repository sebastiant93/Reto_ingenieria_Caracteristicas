#14) Calcula la correlación lineal

#15) Seleccionar variables con correlación mayor o igual a 0.1

Selecciona solo aquellas variables que tengan una correlación mayor o igual a 0.1, dejándolas en otro frame llamado EmpleadosAttritionFinal. No olvides mantener la variable de salidaAttrition; esto es equivalente a borrar las que no cumplen con el límite.

```python
#Se mantienen las variables solicitadas, con el valor absoluto de la correlación.

cols_remaining=np.where(mat_corr["Attrition"].abs()>=0.1)[0] #Se determinan las posiciones de las variables que cumplen la condición

#La variable Attrition quedará almacenada porque presentará correlación=1.0

#Se imprimen las características que se mantendrán a partir del índice de la matriz de correlación
print(f"Las variables que se mantendrán son : \n {list(mat_corr.index[cols_remaining])}")

#Se crea el dataframe solicitado, usando las características encontradas
EmpleadosAttritionFinal=EmpleadosAttrition[list(mat_corr.index[cols_remaining])]

#Para que quede ordenado pasamos la variable de salida "Attrition" al final del dataframe
col_attrition=EmpleadosAttritionFinal.pop("Attrition")
EmpleadosAttritionFinal["Attrition"]=col_attrition.values

#Para terminar, se imprime una muestra del dataframe resultante
EmpleadosAttritionFinal.head(2)

```

#16) Crear variable llamada EmpleadosAttritionPCA

Crea una nueva variable llamada EmpleadosAttritionPCA formada por los componentes principales del frame EmpleadosAttritionFinal. Recuerda que el resultado del proceso PCA es un numpy array, por lo que, para hacer referencia a una columna, por ejemplo, la 0, puedes usar la instrucción EmpleadosAttritionPCA[:,0]).

```python
#Se realiza el proceso solicitado de PCA
#Para evitar posibles resultados incoherentes estandarizo los datos antes de la transformada
from sklearn.preprocessing import StandardScaler

standard_data=StandardScaler().fit_transform(EmpleadosAttritionFinal)

pca_transform=PCA(standard_data.shape[1]) #Se almacena el objeto PCA con la cantidad de columnas obtenidas del dataframe EmpleadosAttritionFinal
pca_transform.fit(standard_data)
EmpleadosAttritionPCA=pca_transform.transform(standard_data) #Se realiza la transformación de los datos

#Se imprime la muestra del resultado obtenido
EmpleadosAttritionPCA
```

#17) Explicar el 80% de la varianza

Agrega el mínimo número de Componentes Principales en columnas del frame EmpleadosAttritionPCA que logren explicar el 80% de la varianza, al frame EmpleadosAttritionFinal. Puedes usar la instrucción assign, columna por columna, llamando a cada unaC0, C1, etc., hasta las que vayas a agregar.

```python
#Como primer paso, y para entender mejor la relevanca de cada característica, calculo el ratio de varianza, y lo almaceno en un dataframe con el nombre de las características
EmpleadosAttrition_VarRatio=pd.DataFrame([pca_transform.explained_variance_ratio_])
EmpleadosAttrition_VarRatio.columns=EmpleadosAttritionFinal.columns
EmpleadosAttrition_VarRatio
```

```python
#Se usa una estructura while para almacenar el número mínimo de variables que explican el 80% de la varianza

#Se crean algunas variables para iterar en el dataframe
var_sum=0 #Suma acumulada
list_of_vars=[] #Características a almacenar
iter=0 # Contador para recorrer dataframe
while var_sum<0.8: #Se suman los valores hasta cuando el resultado es mayor o igual a 0.8
    list_of_vars.append(EmpleadosAttrition_VarRatio.columns[iter]) #Se almacena el nombre de la característica
    var_sum=var_sum+EmpleadosAttrition_VarRatio.iloc[0,iter] #Se acumula la suma
    iter=iter+1 #Se pasa a la siguiente posición del dataframe

print(f"El grupo de variables representativas escogidas que suman el {var_sum}% de los datos son: {list_of_vars}")
```
